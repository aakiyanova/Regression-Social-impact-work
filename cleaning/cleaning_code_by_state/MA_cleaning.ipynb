```python
fiscal = pd.read_sas('../../data/fiscal2018', format = 'sas7bdat', encoding='iso-8859-1')
```

```python
fiscal_names = list(fiscal[fiscal['STNAME'] == 'Massachusetts']['NAME'])
```

```python
MA = pd.read_excel('../../data/state_data_raw/massachusetts.xlsx')
```

```python
MA.head()
```

```python
MA.columns = MA.iloc[0]
MA = MA[1:]
```

```python
MA_fiscal_rename = {'City on a Hill Charter Public School Circuit Street (Distric' : 'City on a Hill Charter Public School Circuit Street (District)',
'City on a Hill Charter Public School Dudley Square (District' : 'City on a Hill Charter Public School Dudley Square (District)',
'Edward M. Kennedy Academy for Health Careers (Horace Mann Ch' : 'Edward M. Kennedy Academy for Health Careers (Horace Mann Charter) (District)',
'Greenfield' : 'Greenfield Commonwealth Virtual District',
'Phoenix Academy Public Charter High School Springfield (Dist' : 'Phoenix Academy Public Charter High School Springfield (District)'}
```

```python
fiscal = fiscal.replace(MA_fiscal_rename)
```

```python
MA['# in Cohort'] = MA['# in Cohort'].astype(str).str.replace(',', '')
MA['# in Cohort'] = pd.to_numeric(MA['# in Cohort'].str.strip(), errors='coerce')
MA['% Graduated'] = pd.to_numeric(MA['% Graduated']) / 100.
```

```python
MA = MA[['District Name', '# in Cohort', '% Graduated']]
```

```python
MA.columns = ['District Name', 'Total', 'Graduation Rate']
```

```python
MA_merged = pd.merge(fiscal[fiscal['STABBR'] == 'MA'], MA, how='inner', left_on='NAME', right_on='District Name')
```

```python
MA_merged.to_csv('../../data/state_data_merged/MA.csv', index=False)
```
