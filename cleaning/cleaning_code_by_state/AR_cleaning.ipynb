Examine data types and missing values.

```python
AR.info()
```

```python
AR = AR[['District_Name', 'Overall Expected Grad', 'Overall Graduation Rate']]
AR
```

```python
AR.columns = ['District Name', 'Total', 'Graduation Rate']
```

Change data types.

```python
AR['Graduation Rate'] = AR['Graduation Rate'].astype(str).str.replace('>', '')
AR['Graduation Rate'] = AR['Graduation Rate'].astype(str).str.replace('%', '')
AR['Graduation Rate'] = AR['Graduation Rate'].astype(str).str.replace('N<10', '')

AR['Graduation Rate'] = pd.to_numeric(AR['Graduation Rate']) / 100
```

```python
AR['Total'] = AR['Total'].astype(str).replace('RV', '')
AR['Total'] = AR['Total'].astype(str).str.replace('N<10', '')
AR['Total'] = pd.to_numeric(AR['Total'])
```

```python
AR = AR.reset_index(drop=True)
```

```python
AR.info()
```

Merge federal and state data, keeping only matches between the two. 

```python
AR_merged = pd.merge(fiscal_AR[fiscal_AR['STABBR'] == 'AR'], AR, how='inner', left_on='NAME', right_on='District Name')
```
