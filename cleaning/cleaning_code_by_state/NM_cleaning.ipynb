```python
abbr = 'NM'
```

```python
file = 'new_mexico2018.xlsx'
```

Read in federal level data

```python
state_grads
```

Filter results.

```python
state_grads = state_grads[(state_grads['School'] == 'Districtwide') &
                         (state_grads['Group'] == 'All Students')]
```

```python
state_grads
```

```python
state_grads['Total'] = np.full_like(state_grads['Group'], '')
```

```python
state_grads.columns = ['District Name', 'Total', 'Graduation Rate']
```

```python
state_grads.info()
```

Convert data types.

```python
state_grads = state_grads.replace('*', '')
```

```python
state_grads['Total'] = pd.to_numeric(state_grads['Total'])
state_grads['Graduation Rate'] = pd.to_numeric(state_grads['Graduation Rate']) / 100
```
