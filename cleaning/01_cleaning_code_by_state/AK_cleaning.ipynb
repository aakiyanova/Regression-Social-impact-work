Names aren't lining up at all, but I'll see if I can match most of them up with some batch editing. 

```python
fiscal_AK = fiscal[fiscal['STNAME'] == 'Alaska']
fiscal_AK.head()
```

```python
fiscal_AK['NAME'] = fiscal_AK['NAME'].astype(str).str.replace(' School District', '')
fiscal_AK['NAME'] = fiscal_AK['NAME'].astype(str).str.replace(' Borough', '')
fiscal_AK['NAME'] = fiscal_AK['NAME'].astype(str).str.replace(' City', '')
fiscal_AK['NAME'] = fiscal_AK['NAME'].astype(str).str.replace(' Public', '')
fiscal_AK['NAME'] = fiscal_AK['NAME'].astype(str).str.replace(' Schools', '')
```

```python
fiscal_AK.head()
```

```python
AK['District'] = AK['District'].astype(str)
```

```python
[name for name in list(AK['District']) if name not in list(fiscal_AK['NAME'])]
```

```python
[name for name in list(fiscal_AK['NAME']) if name not in list(AK['District'])]
```

Replace remaining names with appropriate formattings.

```python
AK_fiscal_rename = {
 'Mount Edgecumbe High School Agency' : 'Mt Edgecumbe',
 'Delta-Greely' : 'Delta/Greely',
 'Ketchikan Gateway' : 'Ketchikan',
 'Matanuska-Susitna' : 'Mat-Su',
 'Iditarod Area' : 'Iditarod',
 'Fairbanks North Star' : 'Fairbanks'
}
```

```python
fiscal_AK = fiscal_AK.replace(AK_fiscal_rename)
```

Examine data types and missing values.

```python
AK.info()
```

Change column names for consistency across states. 

```python
AK = AK[['District', 'Cohort', 'Rate']]
AK.columns = ['District Name', 'Total', 'Graduation Rate']
```

```python
AK
```

Drop the districts with 0 students for the year. 

```python
AK = AK[AK['Total'] > 0]
```

Change data types.

```python
AK['Graduation Rate'] = AK['Graduation Rate'].astype(str).str.replace('%', '')
```

```python
AK['Graduation Rate'] = pd.to_numeric(AK['Graduation Rate']) / 100
```

```python
AK = AK.reset_index(drop=True)
```

Merge federal and state data, keeping only matches between the two. 

```python
AK_merged = pd.merge(fiscal_AK, AK, how='inner', left_on='NAME', right_on='District Name')
```

Save cleaned data. 

```python
AK_merged.to_csv('../../data/state_data_merged/AK.csv', index=False)
```
