```python
import pandas as pd
```

Read in federal level data

```python
fiscal = pd.read_sas('../../data/fiscal2018', format = 'sas7bdat', encoding='iso-8859-1')
```

```python
fiscal
```

Generate list of districts in the state in the federal data

```python
fiscal_names = list(fiscal[fiscal['STNAME'] == 'Alabama']['NAME'])
```

```python
len(fiscal_names)
```

Read in state level data

```python
AL = pd.read_excel('../../data/state_data_raw/alabama2019.xlsx')
```

```python
AL.head()
```

Reset columns.

```python
AL.columns = AL.iloc[3]
AL = AL[4:]
```

Filter for only the district level, total population samples.

```python
AL = AL[(AL['System Name'] == AL['School Name']) &
        (AL['Grade'] == 'Grade 12') &
        (AL['Gender'] == 'All Gender') &
        (AL['Race'] == 'All Race') &
        (AL['Ethnicity'] == 'All Ethnicity') &
        (AL['SubPopulationName'] == 'All SubPopulation')]
```

```python
AL.head()
```

```python
len(AL)
```

Check for non-matches in the two lists

```python
[name for name in list(AL['System Name']) if name not in fiscal_names]
```

```python
[name for name in fiscal_names if name not in list(AL['System Name'])]
```

Examine data types and missing values.

```python
AL.info()
```

```python
AL = AL[['System Name', 'Total', 'Graduation Rate']]
AL
```

Remove duplicate.

```python
AL = AL.iloc[:-1]
```

Alabama source indicates that '~' is for graduation rates >= 95% or <= 5%. Because this represents about half of the data, we can't just drop these samples. We'll assume that very few, if none, of the districts have graduation rates below 5% and instead assign them the minimum value of 95%.

```python
len(AL[AL['Graduation Rate'] == '~'])
```

```python
AL['Graduation Rate'] = AL['Graduation Rate'].replace('~', .95)
```

Change data types.

```python
AL['Graduation Rate'] = pd.to_numeric(AL['Graduation Rate'])
```

```python
AL['Total'] = pd.to_numeric(AL['Total'])
```

```python
AL = AL.reset_index(drop=True)
```

Change column names for consistency across states. 

```python
AL = AL.rename(columns={'System Name' : 'District Name'})
```

Merge federal and state data, keeping only matches between the two. 

```python
AL_merged = pd.merge(fiscal[fiscal['STABBR'] == 'AL'], AL, how='inner', left_on='NAME', right_on='District Name')
```

Save cleaned data. 

```python
AL_merged.to_csv('../../data/state_data_merged/AL.csv', index=False)
```
