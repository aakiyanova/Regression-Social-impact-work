```python
import pandas as pd
import numpy as np
```

```python
abbr = 'IA'
```

Read in federal level data

```python
fiscal = pd.read_sas('../../data/fiscal2018', format = 'sas7bdat', encoding='iso-8859-1')
```

Generate list of districts in the state in the federal data

```python
state_fiscal = fiscal[(fiscal['STABBR'] == abbr) & (fiscal['GSHI'] == '12')]
```

```python
len(state_fiscal)
```

```python
state_fiscal.head()
```

Read in state level data

```python
state_grads = pd.read_excel('../../data/state_data_raw/iowa2018.xlsx')
```

```python
state_grads
```

Reset columns and rows.

```python
state_grads.columns = state_grads.loc[2]
```

```python
state_grads = state_grads.iloc[3:-1, :8]
```

```python
state_grads
```

Select and rename columns.

```python
state_grads = state_grads[['District Name', 'Denominator', 'Rate']]
```

```python
state_grads.columns = ['District Name', 'Total', 'Graduation Rate']
```

```python
state_grads
```

```python
state_grads.info()
```

Convert data types.

```python
state_grads = state_grads.replace('****', '')
```

```python
state_grads['Total'] = pd.to_numeric(state_grads['Total'])
state_grads['Graduation Rate'] = pd.to_numeric(state_grads['Graduation Rate']) / 100
```

Check for matches and non-matches in the two lists

```python
Matches = [name for name in list(state_grads['District Name']) if name in list(state_fiscal['NAME'])]
Matches.sort()
len(Matches)
```

```python
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' Community School District', '')
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' Comm School District', '')
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' School District', '')
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' Sch Dist', '')
state_fiscal['NAME'] = state_fiscal['NAME'].str.replace(' CSD', '')
```

```python
A = [name for name in list(state_grads['District Name']) if name not in list(state_fiscal['NAME'])]
A.sort()
A
```

```python
B = [name for name in list(state_fiscal['NAME']) if name not in list(state_grads['District Name'])]
B.sort()
B
```

Replace the names I can find matches for.

```python
state_grads_rename = {
    'Decorah' : 'Decorah Community',
    'Eddyville-Blakesburg- Fremont' : 'Eddyville-Blakesburg-',
    'Estherville Lincoln Central Com' : 'Estherville Lincoln',
    'Exira-Elk Horn- Kimballton Comm' : 'Exira-Elk Horn-',
    'North Fayette' : 'North Fayette Valley',
    'Odebolt-Arthur' :'Odebolt Arthur Battle Creek Ida Grove'
}
```

```python
state_fiscal = state_fiscal.replace(state_grads_rename)
```

Merge federal and state data, keeping only matches between the two. 

```python
state_grads_merged = pd.merge(state_fiscal, state_grads, how='inner', left_on='NAME', right_on='District Name')
```

Save cleaned data. 

```python
state_grads_merged.to_csv('../../data/state_data_merged/' + abbr + '.csv', index=False)
```
