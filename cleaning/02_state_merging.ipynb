```python
import pandas as pd
import numpy as np

from os import listdir
```

```python
path = '../data/state_data_merged/'
```

```python
listdir(path)
```

```python
states_csv = [file for file in listdir(path) if '.csv' in file]
```

```python
states_df = [pd.read_csv(path + file) for file in states_csv]
```

```python
full_df = pd.concat(states_df)
```

```python
full_df
```

```python
full_df.to_csv('../data/' + str(len(states_df)) + '_states_merged.csv', index=False)
```
